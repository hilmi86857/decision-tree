```python
import pandas as pd
import requests

# Ganti URL berikut dengan tautan *Raw* Gist
url = "https://gist.githubusercontent.com/pravalliyaram/5c05f43d2351249927b8a3f3cc3e5ecf/raw/Mall_Customers.csv"
res = requests.get(url, allow_redirects=True)


# Simpan file ke disk di Colab
with open('Mall_Customers.csv', 'wb') as f:
    f.write(res.content)

# Cek file dan load ke DataFrame
df = pd.read_csv('Mall_Customers.csv')
df.head()

```

```python
# Pilih kolom
customer_data = df[['Annual Income (k$)', 'Spending Score (1-100)']]
customer_data.columns = ['Annual Income', 'Spending Score']

# Tampilkan preview
customer_data.head()


```

```python
from sklearn.cluster import KMeans

inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(customer_data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.grid(True)
plt.show()

```

```python
# Step 5: Training model K-Means dengan jumlah cluster optimal (misal k=4)
kmeans = KMeans(n_clusters=4, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(customer_data)

# Output
customer_data.head()
```
